````
def calculateDerivative(func, x1):
    deltaX = 0.0000001;
    derivative = (func(x1 + deltaX) - func(x1)) / deltaX;
    answerTuple = (derivative, x1, func(x1), deltaX);
    return answerTuple
````

````
def sampleFunc(x) :
    return x*x*x; 
````

````
print(calculateDerivative(sampleFunc, 100))
````

````
def exponentialList(n, x): # In Q., asked for tuple. Can typecast while returning final answer.
    if(n == 0):
        answerList = [1];
        return answerList;
    answerList = exponentialList(n - 1, x);
    nextValue = answerList[-1] + (x**(n))/(math.factorial(n));
    answerList.append(nextValue);
    return answerList;
````

````
encodedString = '';
for char in sampleString:
    encodedString = encodedString + chr(2*ord(char) + 3);
print(encodedString);
````

````
decodedString = '';
for char in encodedString:
    decodedString += chr(int((ord(char) - 3)/2));
print(decodedString);
````
