```python
def draw_img(img, Y):
    # Reshaping so that can be plotted as 2-d image
    plt.imshow(img.reshape(28, 28), label = Y, cmap = 'gray')
    plt.title('Label:' +str(Y))
```

```python
plt.figure(figsize = (18, 18))
for i in range(25):
    curr_X = X[i + 1]; 
    curr_Y = Y[i + 1];
#     Subplotting starts from 1, not 0
    plt.subplot(5, 5, i + 1)
    draw_img(curr_X, curr_Y)
```

# Data Acquisition/Scrapping

```python
# 1. Get HTTP Response
# 2. Get HTML page from it
# 3. Parse the HTML
```

```python
from urllib.request import urlopen
```

```python
dataLink = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area"
```

```python
httpResponse = urlopen(dataLink);
```

```python
print(httpResponse)
```

```python
type(httpResponse)
```

```python
# Reads html code
htmlData = httpResponse.read()
```

```python
htmlData
```

```python
# Closing the connection made through query
httpResponse.close()
```

```python
# This is not a structured html
type(htmlData)
```

```python
# Hence, parsing html data using bs4
# bs4 can be used to parse in other languages as well, like json etc
from bs4 import BeautifulSoup as soup
```

```python
data_soup = soup(htmlData, 'html.parser')
```

```python
data_soup
```

```python
type(data_soup)
```

```python
# Looks like gives only the 1st h1 tag
h1 = data_soup.h1
```

```python
print(h1, type(h1))
```

```python
h2 = data_soup.h2
print(h2)
```

```python
div = data_soup.div
print(div)
```

```python
h2_tags = data_soup.find_all('h2', {})
h2_tags
```

```python
print(type(h2_tags))
```

```python
# Gets us all the tables
tables = data_soup.find_all('table', {})
```

```python
tables
```

```python
tables[1]
```

```python
# To solve this problem of choosing the particular table, we use dictionary above
table = data_soup.find_all('table', {'class':'wikitable sortable'})
table
```

```python
print(type(data_soup))
print(type(tables))
print(type(table[0]))
```

```python
# Doing tables[0] because find_all works on <class 'bs4.BeautifulSoup'> and <class 'bs4.element.Tag'> only
headings = table[0].find_all('th', {})
```

```python
print(headings)
```
