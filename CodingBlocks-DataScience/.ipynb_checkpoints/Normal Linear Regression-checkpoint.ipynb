```python
# Adding noise to y, to create realistic signal
trueValues = trueValues + random_noise;
```

```python
plt.scatter(np.arange(100),trueValues)
```

```python
# Let this be the input values
inputValues = np.arange(100); 
```

```python
# Initializing weights to 0
theta_0 = 0;
theta_1 = 0;
```

```python
# Hypothesis function
hypothesisFunction = theta_1*inputValues + theta_0
# These are the weights that need to be found out
```

```python
# Loss Function
lossFunction = 0.5*np.sum(((hypothesisFunction - trueValues)**2));
```

```python
# We need to reduce this loss
lossFunction
```

```python
# Let us first find theta_1 by assuming theta_0 = 0, and equating derivative of this loss equal to 0.
# This gives an equation(written in copy), from that we can get theta_1
# For this, we need 3 values, sum of all squares of input values, sum of all input values, and sum of product of input and output values.
sumOfInputSquares = np.sum(inputValues**2);
sumOfInputs = np.sum(inputValues);
sumOfProducts = np.sum(inputValues*trueValues);
theta_1 = (sumOfProducts - theta_0*sumOfInputs)/(sumOfInputSquares);
```

```python
theta_1
```

```python
# Now we use theta_1 as constant, and find theta_0 by equating derivative of loss w.r.t. theta_0, to 0.
sumOfTrueValues = np.sum(trueValues);
# 1000 - no.of input values
theta_0 = (sumOfTrueValues - theta_1*(sumOfInputs))/1000; 
```

```python
theta_0
```

```python
# We will run the iterations for these many steps
theta_0 = 0;
theta_1 = 0;
numOfSteps = 10000;
for i in range(numOfSteps):
    theta_1 = (sumOfProducts - theta_0*sumOfInputs)/(sumOfInputSquares);
    theta_0 = (sumOfTrueValues - theta_1*(sumOfInputs))/1000; 
    
```

```python
print(theta_0, theta_1)
```

```python
predictedValues = theta_1*inputValues + theta_0;
```

```python
plt.scatter(np.arange(100), trueValues, label = "line 2")
plt.plot(predictedValues, label = "line 1", color = 'red')
plt.legend()
plt.show()
```

```python
# Converting 1-d array to 2-d array.
# Standardizing values
# Row wise mean
inputMeans = np.mean(inputValues, 1)
inputStds = np.std(inputValues, 1)
trueMean = np.mean(trueValues, 0)
trueStd = np.std(trueValues, 0);
standInput = (inputValues - inputMeans)/inputStds;
standTrue = (trueValues - trueMean)/trueStd;
# Ones added to last row.
standInput = np.vstack((standInput, np.ones(inputValues.shape[1])))
```

```python
# Linear Regression

def getCumulativeHypothesisFunction(thetas, inputValues):
        hypothesisFun = 0;
        for i in range(thetas.shape[0]):
            hypothesisFun += thetas[i]*inputValues[i, :];
        return hypothesisFun;

def getGradient(thetas, inputValues, trueValues, thetaIndex):
        lossDerivative = np.sum((getCumulativeHypothesisFunction(thetas, inputValues) - trueValues)*(inputValues[thetaIndex, :]));
        return lossDerivative;
    
thetas = np.zeros((standInput.shape[0])).reshape(standInput.shape[0],1)
alpha = 0.001;
epochs = 1000;
# Total Loss
loss = [];
for i in range(epochs):
#     Iterate over all partial derivatives of loss functions, updating the value of theta.
    for j in range(standInput.shape[0]):
        lossDerivative = getGradient(thetas, standInput, standTrue, j);
        thetas[j] = thetas[j] - alpha * lossDerivative;
    loss = np.append(loss, (np.sum((thetas * standInput- standTrue)**2))/standInput.shape[1]);
#     print((np.sum((thetas * standInput- standTrue)**2)))
```

```python
import numpy as np;
import matplotlib.pyplot as plt;

class linearRegressionModel:
    def __init__(self, inputValues, trueValues, epochs = 1000, alpha = 0.01):
        self.epochs = epochs;
        self.alpha = alpha;
        self.inputValues = inputValues;
        self.trueValues = trueValues;
        self.standInput = self.prepareInputValue(inputValues);
        self.thetas = self.prepareThetas();
        self.standTrue = self.prepareTrueValue(trueValues);
        self.lossArray = [];

    def prepareInputValue(self, values):
        mean = np.mean(values, 1).reshape((values.shape[0], 1));
        stdDev = np.std(values, 1).reshape((values.shape[0], 1));
        standInput =  (values - mean)/stdDev;
        # Ones added to last row.
        standInput = np.vstack((standInput, np.ones(inputValues.shape[1])));
        return standInput;

    def prepareTrueValue(self, values):
        mean = np.mean(values);
        stdDev = np.std(values);
        return (values - mean)/stdDev;

    def prepareThetas(self):
        # Reshaping to convert from (n,) -> (n, 1) (To avoid multiplication errors)
        return np.zeros((self.standInput.shape[0])).reshape(self.standInput.shape[0], 1);

    # Hypothesis function gives value for just 1 point.
    # This function sums up all hypothesis functions for all values.
    def getCumulativeHypothesisFunction(self, thetas, inputValues):
        hypothesisFun = 0;
        for i in range(thetas.shape[0]):
            hypothesisFun += thetas[i]*inputValues[i, :];
        return hypothesisFun;

    def getGradient(self, thetas, inputValues, trueValues, thetaIndex):
        lossDerivative = np.sum((self.getCumulativeHypothesisFunction(thetas, inputValues) - trueValues)*(inputValues[thetaIndex, :]));
        return lossDerivative;
    
#     Returns average loss.
    def getLoss(self, thetas, inputValues, trueValues):
        return np.sum(((thetas * inputValues - trueValues)**2)/inputValues.shape[1])

    def performGradientDescent(self):
        for i in range(self.epochs):
            for j in range(self.standInput.shape[0]):
                lossDerivative = self.getGradient(self.thetas, self.standInput, self.standTrue, j);
                self.thetas[j] = self.thetas[j] - self.alpha * lossDerivative;
            self.lossArray = np.append(self.lossArray, self.getLoss(self.thetas, self.standInput, self.standTrue));
    
    def plotLoss(self):
        plt.plot(self.lossArray);

```

```python
trueValues = np.arange(10) + 2*np.random.rand(10);
inputValues = np.array([np.arange(10), np.arange(10), np.arange(10), np.arange(10), np.arange(10)]);
lrm = linearRegressionModel(inputValues, trueValues, 1000, 0.001);
```

```python
lrm.performGradientDescent()
```

```python
lrm.plotLoss()
```
